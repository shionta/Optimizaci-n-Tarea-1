```python
# MEZCLA DE FERTILIZANTES - MAXIMIZAR GANANCIA

import numpy as np
from scipy.optimize import minimize

# Función objetivo: Maximizar Z = 180*x1 + 220*x2
# Como minimize busca minimizar, se usa el negativo de la ganancia
def ganancia_negativa(x):
    return -(180*x[0] + 220*x[1])

# Se tienen las siguientes estricciones de desigualdad
# 3*x1 + 4*x2 <= 120
# 2*x1 + 1*x2 <= 80
# 5*x1 + 3*x2 <= 150

def restriccion_A(x):
    return 120 - (3*x[0] + 4*x[1])
def restriccion_B(x):
    return 80 - (2*x[0] + 1*x[1])
def restriccion_C(x):
    return 150 - (5*x[0] + 3*x[1])

# Configurar restricciones en formato scipy
restricciones = [
    {'type': 'ineq', 'fun': restriccion_A},
    {'type': 'ineq', 'fun': restriccion_B},
    {'type': 'ineq', 'fun': restriccion_C}
]

# Límites: x1 >= 0, x2 >= 0
limites = [(0, None), (0, None)]

# Punto inicial
x0 = [10, 10]

# Resolver
resultado = minimize(
    ganancia_negativa,
    x0,
    method='SLSQP',
    bounds=limites,
    constraints=restricciones
)

# Resultados
x1_opt, x2_opt = resultado.x
ganancia_maxima = -resultado.fun

print("=" * 45)
print("Mezcla de fertilizantes - Maximizar ganancia")
print("=" * 45)

print(f"Estado: {'Óptimo encontrado' if resultado.success else 'Fallo'}")
print()
print(f"Fertilizante F1 (x1) = {x1_opt:.2f} toneladas")
print(f"Fertilizante F2 (x2) = {x2_opt:.2f} toneladas")
print()
print(f"Ganancia máxima = ${ganancia_maxima:.2f}")
print()

# Verificación de restricciones
print("Verificación de restricciones:")
print(f"Compuesto A: 3*x1 + 4*x2 = {3*x1_opt + 4*x2_opt:.2f} <= 120")
print(f"Compuesto B: 2*x1 + 1*x2 = {2*x1_opt + 1*x2_opt:.2f} <= 80")
print(f"Compuesto C: 5*x1 + 3*x2 = {5*x1_opt + 3*x2_opt:.2f} <= 150")
print("=" * 45)
```

```python
# PRODUCCIÓN SEMANAL DE UNA FABRICA
# Entrada de datos
S_actual = float(input("S_0 (sillas en la semana 0): "))
M_actual = float(input("M_0 (mesas en la semana 0): "))
semanas = int(input("Número de semanas a evaluar: "))

print(f"Semana 0: Sillas = {S_actual:.2f}, Mesas = {M_actual:.2f}")

# Simulación semana a semana
for n in range(1, semanas + 1):
    S_siguiente = 0.6 * S_actual + 0.2 * M_actual + 40
    M_siguiente = 0.1 * S_actual + 0.5 * M_actual + 20

    print(f"Semana {n}: Sillas = {S_siguiente:.2f}, Mesas = {M_siguiente:.2f}")

    # Actualizar los datos para la siguiente iteración
    S_actual = S_siguiente
    M_actual = M_siguiente
```

```python
# MODELO LOGISTICO DE PESCA
# Entrada de datos
p_actual = float(input("Población inicial P_0: "))
años = int(input("Número de años: "))

#Parametros
r = 0.4
K = 1000
H = 80

print(f"Población inicial año 0= {p_actual:.2f} peces")

# Simulación año a año
for n in range(1, años + 1):
    p_siguiente = p_actual + r * p_actual * (1 - p_actual / K) - H
    print(f"Población en el año {n}= {p_siguiente:.2f} peces")

    # Actualizar los datos para la siguiente iteración
    p_actual = p_siguiente
```
